```python
'''Q1)Which function is used to open a file? What are the different modes of opening a file?
Explain each mode of file opening.

Ans-In Python, the built-in open() function is used to open a file. The open() function returns a file object, which can be used to perform various operations on the
file, such as reading or writing data.

Here are the different modes of opening a file:

'r' (Read mode): This is the default mode. It is used to open a file for reading. 
If the file does not exist, it will raise a FileNotFoundError exception.

'w' (Write mode): This mode is used to open a file for writing. If the file exists, 
it will be truncated to zero length. If the file does not exist, a new file will be created.

'x' (Exclusive creation mode): This mode is used to open a file for exclusive creation.
If the file already exists, it will raise a FileExistsError exception.

'a' (Append mode): This mode is used to open a file for writing in append mode. 
The file pointer will be at the end of the file. If the file does not exist, a new file 
will be created.

'b' (Binary mode): This mode is used to open a file in binary mode. In this mode, 
the file is read or written as a sequence of bytes. This mode is used for non-text files 
such as images or executables.

't' (Text mode): This mode is used to open a file in text mode. In this mode, the file is
read or written as a sequence of characters. This mode is used for text files.

'+' (Update mode): This mode is used to open a file for both reading and writing (updating).
If the file does not exist, it will raise a FileNotFoundError exception.

Q2)Why close() function is used? Why is it important to close a file?

Ans-1)The close() function is used to close a file that has been opened using the open()
function in Python. It is important to close a file after you are done using it because it
releases any system resources that were associated with the file, such as file descriptors 
or memory buffers.
2)If you do not close a file after you are done using it, the operating system may limit the 
number of files you can open simultaneously, which can cause your program to crash or
behave unexpectedly. Additionally, if your program modifies a file and does not close it
properly, the changes may not be saved to the file.

Q3)Write a python program to create a text file. Write ‘I want to become a Data Scientist’ in that file. Then
close the file. Open this file and read the content of the file.

Ans-# Open the file in write mode
with open("myfile.txt", "w") as file:
    file.write("I want to become a Data Scientist")

# Open the file in read mode
with open("myfile.txt", "r") as file:
    contents = file.read()
    print(contents)
    
Q4)Explain the following with python code: read(), readline() and readlines().

Ans-1)read() method:
The read() method reads the entire contents of a file and returns them as a string. It 
takes an optional parameter that specifies the maximum number of bytes to read from the 
file. If you do not specify a parameter, read() will read the entire file.
Here is an example of using the read() method:
# Open the file in read mode
with open("myfile.txt", "r") as file:
    contents = file.read()
    print(contents)
    
2)readline() method:
The readline() method reads a single line from a file and returns it as a string. If you 
call readline() again, it will read the next line of the file. This method is useful when
you want to read a file line by line.
Here is an example of using the readline() method:
# Open the file in read mode
with open("myfile.txt", "r") as file:
    line1 = file.readline()
    line2 = file.readline()
    print(line1)
    print(line2)
    
3)readlines() method:
The readlines() method reads the entire contents of a file and returns them as a list of 
strings, where each string is a line from the file. This method is useful when you want to 
read a file line by line and store the lines in a list.
Here is an example of using the readlines() method:
# Open the file in read mode
with open("myfile.txt", "r") as file:
    # Read the entire contents of the file as a list of strings
    lines = file.readlines()
    # Print each line to the console
    for line in lines:
        print(line)

Q5)Explain why with statement is used with open(). What is the advantage of using with
statement and open() together?

Ans-In Python, the with statement is used with the open() function to ensure that a file
is closed properly after it has been used. The advantage of using with statement and open() 
together is that it simplifies the code for handling file I/O and eliminates the need to
explicitly close the file.


Q6)Explain the write() and writelines() functions. Give a suitable example.

Ans-1)The write() function is used to write a single string to a file. It takes a single
argument, which is the string to be written to the file. If the file does not exist, it 
will be created. If it does exist, the new data will be appended to the end of the file.

Here's an example of using the write() function:
# Open the file for writing
file = open("myfile.txt", "w")
file.write("I want to become a data scientist")
file.close()

2)The writelines() function is used to write a list of strings to a file. It takes a 
single argument, which is a list of strings to be written to the file. Each string in the
list will be written to a new line in the file. If the file does not exist, it will be
created. If it does exist, the new data will be appended to the end of the file.
Here's an example of using the writelines() function:

'''
```
