```python
'''
Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.
Ans-In object-oriented programming (OOP), a class is a blueprint or template for creating 
objects that have certain characteristics and behaviors. It defines the properties and methods 
that an object of that class will have.

An object is an instance of a class, which means it's a specific realization of that class. 
It has its own state (values of its properties) and behavior (how it responds to method calls).

For example, let's say we want to model a class called Car in OOP. We can define the Car class
with properties such as make, model, year, color, and methods such as start(), stop(), 
accelerate(), and brake().

Q2. Name the four pillars of OOPs.

Ans-The four pillars of object-oriented programming (OOP) are:

Encapsulation: It refers to the practice of bundling data and methods that operate on that data
within a single unit (class). The internal workings of the object are hidden from the outside 
world, and only the public interface is accessible. This helps to protect the data from
unauthorized access and modification and makes the code more modular and reusable.

Inheritance: It allows a new class to be based on an existing class (parent or base class), 
inheriting its properties and methods. The new class (child or derived class) can add new
functionality or modify existing functionality of the base class. This promotes code reuse, 
reduces code duplication, and allows for polymorphism (different objects of the same class can 
behave differently).

Polymorphism: It refers to the ability of objects of different classes to be treated as if they
are of the same class, but still respond to method calls differently. There are two types of
polymorphism: static (compile-time) and dynamic (run-time).

Abstraction: It means representing complex real-world entities as simplified models or concepts 
in the program. It allows us to focus on the essential features of an object and ignore the
irrelevant details. Abstraction can be achieved through abstract classes, interfaces, and other
mechanisms that allow us to define common behaviors and attributes for a group of objects 
without specifying their exact implementation.

Q3. Explain why the __init__() function is used. Give a suitable example.

Ans-The __init__() function is a special method in Python classes that is called when an object
is created. It is used to initialize the attributes (properties) of the object with the values 
passed as arguments. 
The __init__() method is commonly used to set default values for the attributes of the object 
and to perform any necessary setup operations before the object can be used. It can also be
used to validate the input values and raise exceptions if necessary.

Q4. Why self is used in OOPs?

Ans-In Object-Oriented Programming (OOP), self is a reference variable that refers to the 
current instance of the class. It is used to access the attributes (properties) and methods of
the object within the class definition.

When a method is called on an object, Python automatically passes the object as the first 
argument to the method, which is referred to as self. This allows the method to access the
object's attributes and call its methods. By convention, the first parameter of each method in
a class should always be self.

Q5. What is inheritance? Give an example for each type of inheritance.

Ans-Inheritance is a fundamental concept in Object-Oriented Programming (OOP) that allows a new class to be based on an existing class (called the base or parent class) and inherit its properties (attributes and methods). This is useful for creating new classes that are similar to existing ones, but with some modifications or additional functionality. Inheritance promotes code reuse, reduces code duplication, and enables polymorphism (different objects of the same class can behave differently).

In Python, there are three types of inheritance:

1)Single inheritance: In single inheritance, a class inherits from a single parent class.
This is the most common type of inheritance. Here's an example:
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def eat(self):
        print("The animal is eating...")

class Dog(Animal):
    def __init__(self, name, age, breed):
        super().__init__(name, age)
        self.breed = breed

    def bark(self):
        print("Woof! Woof!")

my_dog = Dog("Buddy", 5, "Labrador")
print(f"My dog {my_dog.name} is {my_dog.age} years old and is a {my_dog.breed}.")
my_dog.eat()   # Output: "The animal is eating..."
my_dog.bark()  # Output: "Woof! Woof!"

2)Multiple inheritance: In multiple inheritance, a class inherits from multiple parent classes. 
This allows a child class to combine the features of two or more classes. Here's an example:
class Father:
    def __init__(self, height):
        self.height = height

    def display_height(self):
        print(f"Father's height is {self.height} feet.")

class Mother:
    def __init__(self, weight):
        self.weight = weight

    def display_weight(self):
        print(f"Mother's weight is {self.weight} kgs.")

class Child(Father, Mother):
    def __init__(self, height, weight):
        Father.__init__(self, height)
        Mother.__init__(self, weight)

my_child = Child(6.0, 70)
my_child.display_height()  # Output: "Father's height is 6.0 feet."
my_child.display_weight()  # Output: "Mother's weight is 70 kgs."

3)
```
