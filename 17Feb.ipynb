```python
'''Q1. What is MongoDB? Explain non-relational databases in short. In which scenarios it is
preferred to use MongoDB over SQL databases?

Ans-1)MongoDB is a document-oriented, NoSQL database that uses JSON-like documents with optional
schemas. It is designed to store and manage unstructured data and allows for highly flexible 
and scalable data modeling. MongoDB is open source and has a large and active community of 
developers and users.
2)Non-relational databases, or NoSQL databases, are databases that do not use the traditional 
tabular, relational data model of SQL databases. Instead, NoSQL databases use various data 
models such as document-oriented, key-value, column-family, and graph databases. They are 
designed to handle large volumes of unstructured or semi-structured data and offer high 
scalability and performance.
3)MongoDB is preferred over SQL databases in scenarios where there is a need for:Flexible data
modeling: MongoDB's document-based data model allows for a flexible schema that can easily 
accommodate changes in data structures and relationships without requiring a database schema 
update.
High scalability and performance: MongoDB uses horizontal scaling, which allows it to handle 
large volumes of data by adding more servers to a cluster. It also provides fast read and write
operations.
Unstructured or semi-structured data: MongoDB is designed to handle unstructured data such as
documents, images, videos, and social media data.
Real-time data processing: MongoDB supports real-time data processing, making it suitable for
use cases that require real-time analytics and processing, such as online gaming, e-commerce, 
and IoT.
Cloud-based applications: MongoDB is well-suited for cloud-based applications as it can be 
easily deployed on cloud platforms like AWS, Google Cloud, and Microsoft Azure.

Q2.State and Explain the features of MongoDB.

Ans-MongoDB is a popular NoSQL document-oriented database that has a number of features that 
make it a popular choice for developers and enterprises. Some of the key features of MongoDB are:
1)Document data model: MongoDB uses a document data model, where data is stored in flexible, JSON-like
documents that can have varying structures and fields. This allows for easy and flexible data 
modeling and supports nested and hierarchical data structures.
2)Dynamic schema: MongoDB has a dynamic schema, which means that documents in a collection do not 
have to have the same fields or structure. This allows for easy and flexible data modeling and 
supports nested and hierarchical data structures.
3)High availability: MongoDB supports automatic replication and failover, which means that it can
maintain multiple copies of data across multiple nodes in a cluster. If one node fails, another
node can take over automatically without any disruption to the application.
4)Horizontal scalability: MongoDB supports horizontal scaling, which allows it to handle large 
amounts of data by adding more nodes to a cluster. This makes it easy to scale up and down as 
needed.
5)Rich query language: MongoDB provides a rich query language that supports a wide range of
operations, including queries, updates, and aggregation. It also supports full-text search and 
geospatial queries.
6)Indexing: MongoDB supports a wide range of indexing options, including single field, compound,
multikey, and geospatial indexes. This makes it easy to optimize queries and improve performance.
7)Native aggregation: MongoDB provides native support for aggregation, which allows for complex 
data analysis and processing. This can help improve performance and reduce data processing times.
8)GridFS: MongoDB provides GridFS, a specification for storing and retrieving large files, such 
as images, videos, and audio files. This allows for efficient and scalable storage of large 
files.

Q3.Write a code to connect MongoDB to Python. Also, create a database and a collection in MongoDB.

Ans-import pymongo
client = pymongo.MongoClient('mongodb://localhost:27017/')
db = client['mydatabase']
collection = db['mycollection']
post = {"title": "My first post", "content": "This is my first blog post!", "author": "John Doe"}
result = collection.insert_one(post)
print("Inserted post with ID: {}".format(result.inserted_id))

Q4.Using the database and the collection created in question number 3, write a code to insert one record,
and insert many records. Use the find() and find_one() methods to print the inserted record.

Ans-import pymongo

# Establishing a connection to MongoDB
client = pymongo.MongoClient('mongodb://localhost:27017/')

# Creating a database
db = client['mydatabase']

# Creating a collection
collection = db['mycollection']

# Inserting one record into the collection
post = {"title": "My second post", "content": "This is my second blog post!", "author": "Jane Doe"}
result = collection.insert_one(post)
print("Inserted post with ID: {}".format(result.inserted_id))

# Inserting many records into the collection
posts = [
    {"title": "My third post", "content": "This is my third blog post!", "author": "John Smith"},
    {"title": "My fourth post", "content": "This is my fourth blog post!", "author": "Jane Smith"},
    {"title": "My fifth post", "content": "This is my fifth blog post!", "author": "John Doe"},
]
result = collection.insert_many(posts)
print("Inserted posts with IDs: {}".format(result.inserted_ids))

# Finding one record from the collection
post = collection.find_one({"author": "John Doe"})
print("Found one post: {}".format(post))

# Finding all records from the collection
posts = collection.find({})
for post in posts:
    print("Found post: {}".format(post))
    
Q5.Explain how you can use the find() method to query the MongoDB database. Write a simple code
to demonstrate this.

Ans-The find() method is used to query the MongoDB database and retrieve data from a collection.
It takes an optional filter object as an argument to specify the criteria for matching documents.
The filter object can contain key-value pairs that define the fields and values to match.
Here's an example code that demonstrates how to use the find() method to query a MongoDB database:
import pymongo

# Establishing a connection to MongoDB
client = pymongo.MongoClient('mongodb://localhost:27017/')

# Creating a database
db = client['mydatabase']

# Creating a collection
collection = db['mycollection']

# Inserting some data into the collection
collection.insert_many([
    {"name": "John", "age": 30},
    {"name": "Jane", "age": 25},
    {"name": "Bob", "age": 40},
    {"name": "Alice", "age": 35},
    {"name": "Mark", "age": 45}
])

# Querying the collection using the find() method
result = collection.find({"age": {"$gt": 30}})

# Printing the results
for doc in result:
    print(doc)
    
Q6. Explain the sort() method. Give an example to demonstrate sorting in MongoDB.

Ans-The sort() method in MongoDB is used to sort the results of a query in ascending or
descending order based on one or more fields. The sort() method takes a dictionary as an
argument, where each key represents a field to sort by and the corresponding value is either 1
for ascending order or -1 for descending order.
Here's an example code that demonstrates sorting in MongoDB using the sort() method:
import pymongo

# Establishing a connection to MongoDB
client = pymongo.MongoClient('mongodb://localhost:27017/')

# Creating a database
db = client['mydatabase']

# Creating a collection
collection = db['mycollection']

# Inserting some data into the collection
collection.insert_many([
    {"name": "John", "age": 30},
    {"name": "Jane", "age": 25},
    {"name": "Bob", "age": 40},
    {"name": "Alice", "age": 35},
    {"name": "Mark", "age": 45}
])

# Sorting the collection in ascending order by age
result = collection.find().sort("age", 1)

# Printing the sorted results
for doc in result:
    print(doc)
    
Q7.Explain why delete_one(), delete_many(), and drop() is used.

Ans-In MongoDB, there are different methods available to delete documents or collections from a
database. These methods are delete_one(), delete_many(), and drop().

delete_one() method is used to delete a single document that matches the specified filter criteria. If multiple documents match the filter criteria, only the first one will be deleted.

delete_many() method is used to delete all documents that match the specified filter criteria.

drop() method is used to drop a collection from the database. Dropping a collection means removing all the documents from the collection and deleting the collection itself.
'''
```
