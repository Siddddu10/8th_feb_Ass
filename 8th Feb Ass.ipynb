```python
'''Q1)What is Abstraction in OOps? Explain with an example.

Ans--1)In Object-Oriented Programming (OOPs), Abstraction is the process of hiding the 
implementation details of a particular feature while displaying only the essential
details to the user.
2)It allows programmers to create complex systems by breaking them down into 
smaller, more manageable components.
Eg-Thus, we can create an abstract class called "Car" that defines the essential 
features of a car object, as shown below:
abstract class Car {
  abstract int getCurrentSpeed();
  abstract int getNumberOfDoors();
  abstract String getFuelType();
}

class SportsCar extends Car {
  int getCurrentSpeed() {
    return 200;
  }
  int getNumberOfDoors() {
    return 2;
  }
  String getFuelType() {
    return "Gasoline";
  }
}

class ElectricCar extends Car {
  int getCurrentSpeed() {
    return 150;
  }
  int getNumberOfDoors() {
    return 4;
  }
  String getFuelType() {
    return "Electricity";
  }
}

Q2)Differentiate between Abstraction and Encapsulation.Explain with an example

Ans-1) Abstraction is about focusing on the essential features of an object or 
system and ignoring the details that are not relevant to the user. 
For example, in the case of a car object, the essential features that are
relevant to the user might include the car's current speed, the number of doors,
and the type of fuel it uses. 
The internal details of how the car's engine works, how the transmission shifts
gears, or how the fuel is ignited are not essential details for the user.

Thus, we can create an abstract class called "Car" that defines the essential 
features of a car object, as shown below:
abstract class Car {
  abstract int getCurrentSpeed();
  abstract int getNumberOfDoors();
  abstract String getFuelType();
}

class SportsCar extends Car {
  int getCurrentSpeed() {
    return 200;
  }
  int getNumberOfDoors() {
    return 2;
  }
  String getFuelType() {
    return "Gasoline";
  }
}

class ElectricCar extends Car {
  int getCurrentSpeed() {
    return 150;
  }
  int getNumberOfDoors() {
    return 4;
  }
  String getFuelType() {
    return "Electricity";
  }
}
2)Encapsulation is about protecting an object's internal state and behavior from
outside interference. In the case of a car object, the internal state might 
include the car's current speed, the fuel level, and the engine temperature.
These internal states should not be directly accessed or manipulated by code 
outside the car object. Instead, we can provide public methods that allow 
external code to access and modify the car's internal state in a controlled 
manner.
class Car {
  private int currentSpeed;
  private int fuelLevel;
  private int engineTemperature;

  public void setSpeed(int speed) {
    this.currentSpeed = speed;
  }

  public void setFuelLevel(int fuelLevel) {
    this.fuelLevel = fuelLevel;
  }

  public void setEngineTemperature(int engineTemperature) {
    this.engineTemperature = engineTemperature;
  }

  public int getSpeed() {
    return this.currentSpeed;
  }

  public int getFuelLevel() {
    return this.fuelLevel;
  }

  public int getEngineTemperature() {
    return this.engineTemperature;
  }
}

Q3-What is abc module in python? Why is it used?

Ans-In Python, the abc (Abstract Base Classes) module provides a framework for 
defining abstract base classes. Abstract base classes are classes that are not 
meant to be instantiated, but rather serve as a blueprint for other classes to
inherit from. They define a set of methods and properties that derived classes 
must implement in order to be considered "compatible" with the base class.

The abc module is used to create abstract base classes and enforce interface
specifications in Python. It provides a way to define abstract methods, which 
are methods that must be implemented by any concrete subclass. If a subclass
does not implement all the abstract methods, an error is raised.

Here is an example of how to use the abc module to define an abstract base 
class:

from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass
        
Q4-Can we create an instance of an abstract class? Explain your answer.  

Ans- No, we cannot create an instance of an abstract class in Python. 
An abstract class is a class that is meant to be subclassed and not instantiated
on its own. It is designed to serve as a blueprint for other classes to inherit
from, and it defines a set of methods and properties that derived classes must
implement in order to be considered "compatible" with the base class.

When we try to create an instance of an abstract class, Python raises a TypeError 
at runtime, indicating that we cannot create an instance of an abstract class
'''
```
